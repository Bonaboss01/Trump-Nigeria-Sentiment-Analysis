```python

data_dir = os.path.join(os.getcwd(), "..", "data")
os.makedirs(data_dir, exist_ok=True)

# --- Base configuration ---
num_rows = 2000
start_time = datetime(2025, 11, 2, 8, 0, 0)

# Different narrative buckets
anti_trump_templates = [
    "Trump threatening to come into Nigeria 'guns blazing' is a complete insult to our sovereignty.",
    "You cannot reduce Nigeria's complex crisis to 'saving Christians'. Trump's statement is reckless.",
    "Trump oversimplifies our security issues. This is not Hollywood, it's people's lives.",
    "Violence affects Muslims and Christians. Trump's framing is dishonest and dangerous.",
    "We reject Trump's threat. Nigeria is not a playground for US politics.",
]

pro_trump_templates = [
    "Finally a leader calling out the killing of Christians in Nigeria. Strong words from Trump are needed.",
    "Trump is right to warn Nigeria. Someone has to stand for persecuted Christians.",
    "People blame Trump, but Christians in Nigeria are suffering. At least he is speaking.",
]

sovereignty_templates = [
    "No foreign president has the right to threaten invasion. Nigeria is a sovereign state.",
    "Whatever our internal issues, military threats from Trump are unacceptable.",
    "This language sets a dangerous precedent for USâ€“Africa relations.",
]

neutral_news_templates = [
    "Trump threatens to cut aid and consider action over alleged killings of Christians; Nigerian government rejects the claims.",
    "SUMMARY: Trump issues warning to Nigeria, officials respond, observers call for calm.",
    "Developing story: Trump comments on Nigeria spark diplomatic reactions at home and abroad.",
]

sarcasm_templates = [
    "Trump said 'guns blazing'? Abeg make una no use our country do action film.",
    "Before Trump reaches Nigeria, fuel go finish for plane.",
    "Na every issue Trump dey use Nigeria do content. Rest small.",
]

factcheck_templates = [
    "Reminder: violence in Nigeria is driven by multiple factors, not only attacks on Christians.",
    "Context matters: both Christians and Muslims have been victims. Stop using half-truths.",
    "Media repeating Trump's line without nuance is misleading and harmful.",
]

misc_templates = [
    "We need real reforms, not loud threats from abroad.",
    "Pray for Nigeria, but also hold leaders accountable at home and abroad.",
    "Ordinary Nigerians will suffer the consequences of this rhetoric, not the politicians.",
]

username_pool = [
    "NaijaVoice", "LagosAnalyst", "ArewaCitizen", "AfricaPolicyWatch", "FaithAndFacts",
    "PolicyNerd", "GlobalObserver", "USConservative", "SaveTheChurch", "AbujaInsider",
    "NaijaLaw", "MediaWatchNG", "FactCheckAfrica", "WorldNewsNow", "BreakingAfrica",
    "NaijaBants", "WaffiGuy", "DiplomatDesk", "HumanRightsWatch", "AnalystEU",
    "PastorK", "NorthCentralVoice", "USVoter", "NaijaStudent", "PeaceNow",
]

buckets = [
    ("anti_trump", anti_trump_templates),
    ("pro_trump", pro_trump_templates),
    ("sovereignty", sovereignty_templates),
    ("neutral_news", neutral_news_templates),
    ("sarcasm", sarcasm_templates),
    ("factcheck", factcheck_templates),
    ("misc", misc_templates),
]

rows = []

for i in range(num_rows):
    stance, templates = random.choice(buckets)
    text = random.choice(templates)

    # Slightly vary texts to avoid exact duplicates
    suffix_options = [
        "",
        " #Nigeria",
        " #Trump",
        " #NigeriaDecides",
        " #PrayForNigeria",
        " This is getting serious.",
        " People need to read more before reacting.",
        " Think about the real people affected.",
    ]
    text = text + random.choice(suffix_options)

    # Create a pseudo-random timestamp within a 2-day window
    delta_minutes = random.randint(0, 60 * 48)
    ts = start_time + timedelta(minutes=delta_minutes)

    # Simple like/retweet logic by stance
    if stance in ["anti_trump", "sovereignty", "factcheck", "sarcasm"]:
        likes = random.randint(50, 800)
        retweets = random.randint(10, 300)
    elif stance == "pro_trump":
        likes = random.randint(30, 600)
        retweets = random.randint(5, 200)
    elif stance == "neutral_news":
        likes = random.randint(10, 200)
        retweets = random.randint(5, 100)
    else:  # misc
        likes = random.randint(10, 400)
        retweets = random.randint(3, 150)

    username = random.choice(username_pool)

    rows.append({
        "date": ts.isoformat(),
        "username": username,
        "text": text,
        "likes": likes,
        "retweets": retweets,
    })

df = pd.DataFrame(rows, columns=["date", "username", "text", "likes", "retweets"])
csv_path = os.path.join(data_dir, "tweets_raw.csv")
df.to_csv(csv_path, index=False)


df.head()
```
